# Gender Distribution of Top and Bottom Ten Countries for the year 2022

```python
# Filter for the indicator codes for male and female population
male_population = ds[ds['Indicator Code'] == 'SP.POP.TOTL.MA.IN']
female_population = ds[ds['Indicator Code'] == 'SP.POP.TOTL.FE.IN']

# Extract data for the year 2022
male_population_2022 = male_population[['Country Name', '2022']].rename(columns={'2022': 'Male Population'})
female_population_2022 = female_population[['Country Name', '2022']].rename(columns={'2022': 'Female Population'})

# Merge male and female population data
population_2022 = pd.merge(male_population_2022, female_population_2022, on='Country Name')

# Calculate total population
population_2022['Total Population'] = population_2022['Male Population'] + population_2022['Female Population']

# Identify top 10 countries by total population
top_10_countries = population_2022.sort_values(by='Total Population', ascending=False).head(10)
print(top_10_countries[['Country Name', 'Male Population', 'Female Population', 'Total Population']])
```

```python
for index, row in top_10_countries.iterrows():
    labels = ['Male', 'Female']
    sizes = [row['Male Population'], row['Female Population']]

    plt.figure(figsize=(6, 6))
    plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
    plt.title(f'Gender Distribution in {row["Country Name"]} (2022)')
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    plt.show()
```

```python
bottom_10_countries = population_2022.sort_values(by='Total Population', ascending=True).head(10)
print(bottom_10_countries[['Country Name', 'Male Population', 'Female Population', 'Total Population']])
```

```python
for index, row in bottom_10_countries.iterrows():
    labels = ['Male', 'Female']
    sizes = [row['Male Population'], row['Female Population']]

    plt.figure(figsize=(6, 6))
    plt.pie(sizes, labels=labels,autopct='%1.1f%%', startangle=140)
    plt.title(f'Gender Distribution in {row["Country Name"]} (2022)')
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    plt.show()
```
